```python
## imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV,StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
### create the Dataset
X,y = make_classification(n_samples = 1000,n_features = 10, n_classes = 2,random_state = 20)
```

```python
X
```

```python
y
```

```python
## train-test-split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

```python
## Model Training

logistic = LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
y_pred = logistic.predict(X_test)
y_pred
```

```python
## perforance metrics
Accuracy = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print("The accuracy from the model is " ,+Accuracy)
print(cm)
print(classification_report(y_test,y_pred))
```

Hyperparameter Tuning and Cross Validation

```python
model = LogisticRegression()
c_values = [100,10,1.0,0.1,0.01]
```

```python
params = [
    # Group 1: liblinear
    {'solver': ['liblinear'], 'penalty': ['l1', 'l2'], 'C': c_values},

    # Group 2: saga (supports all penalties)
    {'solver': ['saga'], 'penalty': ['l1', 'l2', 'elasticnet'], 'C': c_values, 'l1_ratio': [0.5]},

    # Group 3: others (only support l2)
    {'solver': ['lbfgs', 'newton-cg', 'sag'], 'penalty': ['l2'], 'C': c_values}
]
```

```python
## Grid search CV
cv = StratifiedKFold()
grid = GridSearchCV(estimator = model,param_grid = params,scoring = 'accuracy',cv =cv,n_jobs = -1)
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
print(grid.best_score_)
```

```python
y_pred = grid.predict(X_test)
```

```python
## perforance metrics using grid search
Accuracy = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print("The accuracy from the model is " ,+Accuracy)
print(cm)
print(classification_report(y_test,y_pred))
```

Randomized SearchCV

```python
model = LogisticRegression()
rand = RandomizedSearchCV(estimator = model,param_distributions = params,cv = 5, scoring ='accuracy',n_jobs = -1)
rand.fit(X_train,y_train)
```

```python
rand.best_params_
```

```python
rand.best_score_
print(rand.best_score_)
```

```python
y_pred = rand.predict(X_test)
```

```python
## perforance metrics using randomized search
Accuracy = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
print("The accuracy from the model is " ,+Accuracy)
print(cm)
print(classification_report(y_test,y_pred))
```
